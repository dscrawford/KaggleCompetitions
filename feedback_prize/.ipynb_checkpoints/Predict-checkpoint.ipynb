```python
def tokenize_text(text_path, df, type_labels):
    doc = df[df['id'] == text_path.split('/')[-1].replace('.txt', '')]
    text = open(text_path).read()
    words = text.split()
    labels = np.full(len(words), type_labels['None'])
    for distype, preds in zip(doc['discourse_type'], doc['predictionstring']):
        indices = [int(i) for i in preds.split()]
        labels[indices] = type_labels[distype]
    return words, labels
```
